### C.Entri dictionary qrels_dict

Kode ini membuat sebuah kamus kosong dengan nama qrels_dict. Selanjutnya, dalam setiap iterasi melalui elemen-elemen dalam all_qrels, code memeriksa apakah query_id sudah ada dalam qrels_dict. Jika iya, maka doc_id ditambahkan ke nilai yang sudah ada dengan dipisahkan oleh koma dan spasi. Jika belum ada, maka doc_id ditambahkan sebagai nilai baru dalam kamus qrels_dict. Dengan melakukan loop ini, kita mengumpulkan doc_id yang terkait dengan setiap query_id dalam kamus qrels_dict.

Kode ini melakukan iterasi melalui setiap pasangan kunci-nilai dalam qrels_dict. Di setiap iterasi, code menggunakan metode replace() untuk mengganti setiap koma dengan koma dan spasi dalam nilai doc_ids. Hal ini dilakukan untuk memastikan ada spasi setelah setiap koma dalam string doc_ids.

Dengan melakukan kedua blok kode ini, kita menghasilkan kamus qrels_dict yang berisi doc_id yang terkait dengan setiap query_id dalam format yang diinginkan.

```python
qrels_dict = {}

for qrel in all_qrels:
    if qrel.query_id in qrels_dict:
        qrels_dict[qrel.query_id] += ', ' + qrel.doc_id
    else:
        qrels_dict[qrel.query_id] = qrel.doc_id

for query_id, doc_ids in qrels_dict.items():
    qrels_dict[query_id] = doc_ids.replace(',', ', ')
```

```python
import nltk
import string

nltk.download('stopwords')
nltk.download('punkt')

from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.lancaster import LancasterStemmer #A
```

### 2.1.Defining Function

Kode dibawah adalah kode fungsi yang digunakan untuk melakukan process stemming

```python
def remove_chars(input_list, chars):
    return [string.translate({ord(char): None for char in chars}) for string in input_list]
```

```python
def proc_stemm(text):
  stoplist = set(stopwords.words('english'))
  st = LancasterStemmer()
  chars = ["/"] #put all punctuation to remove in this list
  word_list = [st.stem(word) for word in remove_chars(word_tokenize(text.lower()), chars) #
  if not word in stoplist and not word in string.punctuation]
  
  return word_list
```

### 3.2.TF

Fungsi get_terms() mengambil teks sebagai input dan menghitung frekuensi kemunculan setiap kata dalam teks tersebut. Frekuensi kata-kata disimpan dalam kamus terms, di mana setiap kata menjadi kunci dan nilai yang sesuai adalah jumlah kemunculannya dalam teks. Fungsi ini mengembalikan kamus terms yang berisi informasi frekuensi kata-kata dalam teks.

```python
def get_terms(text): #A
    terms = {}

    for word in text:
        terms[word] = terms.get(word, 0) + 1
    return terms
```

```python
doc_terms = {}
qry_terms = {}
```

for qry_id in stem_qry.keys(): melakukan hal yang sama seperti di atas, tetapi untuk kamus stem_qry. Setiap teks query yang sesuai dengan qry_id diambil dari kamus stem_qry, diproses menggunakan fungsi get_terms(), dan hasil frekuensi kemunculan kata-kata disimpan dalam kamus qry_terms dengan menggunakan qry_id sebagai kunci.

```python
for doc_id in stem_doc.keys(): #B
    doc_terms[doc_id] = get_terms(stem_doc.get(doc_id))
for qry_id in stem_qry.keys():
    qry_terms[qry_id] = get_terms(stem_qry.get(qry_id))
```

```python
def collect_vocabulary():
    all_terms = []
    for doc_id in doc_terms.keys():
        for term in doc_terms[doc_id].keys():            
            all_terms.append(term)
    for qry_id in qry_terms.keys():
        for term in qry_terms[qry_id].keys():
            all_terms.append(term)
    return sorted(set(all_terms))
```

Sekarang setiap query dan setiap dokumen dapat direpresentasikan dengan sebuah kamus yang memiliki kumpulan kunci yang sama - yaitu term-term dari kosakata bersama. Nilai-nilainya akan sama dengan frekuensi term dalam query atau dokumen tertentu, atau bernilai 0 jika term tersebut tidak ada dalam query atau dokumen.

Fungsi vectorize menangkap fungsionalitas ini. Dengan menggunakan fungsi ini, Anda dapat merepresentasikan semua query dan dokumen dalam sebuah shared space

```python
def vectorize(input_features, vocabulary):
    output = {}
    for item_id in input_features.keys():
        features = input_features.get(item_id)
        output_vector = []
        for word in vocabulary:
            if word in features.keys():
                output_vector.append(int(features[word]))
            else:
                output_vector.append(0)
        output[item_id] = output_vector    
    return output
```

### 3.3.IDF

```python
import math
```

Fungsi calculate_idfs digunakan untuk menghitung nilai IDF (Inverse Document Frequency) dari setiap term dalam kamus vocabulary berdasarkan fitur-fitur dokumen dalam kamus doc_features. Loop pertama iterasi melalui setiap term dalam kamus vocabulary. Pada setiap iterasi, loop kedua menghitung jumlah dokumen yang mengandung term tersebut dari kamus doc_terms, dan kemudian menghitung nilai IDF untuk term tersebut. Hasil nilai IDF dari setiap term disimpan dalam kamus doc_idfs dan kemudian dikembalikan oleh fungsi ini.

```python
def calculate_idfs(vocabulary, doc_features):
    doc_idfs = {}
    for term in vocabulary:
        doc_count = 0
        for doc_id in doc_terms.keys():
            terms = doc_terms.get(doc_id)
            if term in terms.keys():
                doc_count += 1
        doc_idfs[term] = math.log(float(len(doc_terms.keys()))/float(1 + doc_count), 10)
    return doc_idfs
```

### 3.4.Vector representation

Selanjutnya, Anda mendefinisikan sebuah fungsi vectorize_idf untuk menerapkan pembobotan IDF (Inverse Document Frequency) pada struktur data input_terms (khususnya, doc_terms). Secara khusus, dalam fungsi ini Anda mengalikan frekuensi term dengan bobot IDF jika term tersebut ada dalam dokumen; jika tidak, frekuensi term tetap 0.

```python
def vectorize_idf(input_terms, input_idfs, vocabulary):
    output = {}
    for item_id in input_terms.keys():
        terms = input_terms.get(item_id)
        output_vector = []
        for term in vocabulary:
            if term in terms.keys():
              output_vector.append(input_idfs.get(term)*float(terms.get(term)))
            else:
                output_vector.append(float(0))
        output[item_id] = output_vector
    return output
```

### 3.5.Similarity estimation

Mari kita terapkan Cosine Similarity pada query dan dokumen-dokumen dalam dataset dan mengembalikan daftar-dasar dokumen yang relevan berdasarkan skor relevansi mereka, yaitu nilai Cosine Similarity.

Kode di bawah ini menggunakan fungsi itemgetter dari operator, yang membantu mengurutkan kamus Python berdasarkan kunci atau nilai. Kode ini mengimplementasikan tiga fungsi pembantu yang memungkinkan Anda menghitung:

panjang vektor input,
perkalian titik dua vektor, dan
Cosine Similarity antara vektor input yang mewakili, misalnya, sebuah query dan sebuah dokumen.
Anda menginisialisasi query dengan mengekstrak query dengan qry_id tertentu, misalnya, query 3. Kemudian, untuk setiap dokumen dalam kumpulan dokumen, Anda menghitung Cosine Similarity antara query input dan dokumen tersebut, dan menyimpan id dokumen sebagai kunci dan nilai kosinus sebagai nilai dalam kamus hasil.

```python
from operator import itemgetter 
```

```python
def length(vector): 
  sq_length = 0
  for index in range(0, len(vector)):
    sq_length += math.pow(vector[index], 2)
  return math.sqrt(sq_length)
```

```python
def dot_product(vector1, vector2):
  if len(vector1)==len(vector2):
    dot_prod = 0
    for index in range(0, len(vector1)):
      if not vector1[index]==0 and not vector2[index]==0:
        dot_prod += vector1[index]*vector2[index]
    return dot_prod
  else:
    return "Unmatching dimensionality"
```

```python
def calculate_cosine(query, document):
  cosine = dot_product(query, document) / (length(query) * (length(document)+(0.001)))
  return cosine
```

###Calculate Precision

Fungsi calculate_precision digunakan untuk menghitung precision (presisi) dari model_output berdasarkan standar emas (gold_standard). Precision mengukur sejauh mana model berhasil mengidentifikasi item yang benar di antara item yang diprediksi.

Fungsi ini menggunakan variabel true_pos untuk menghitung jumlah item yang ditemukan baik oleh model maupun standar emas. Pertama, kita menginisialisasi true_pos dengan nilai 0. Kemudian, untuk setiap item dalam model_output, kita periksa apakah item tersebut ada dalam gold_standard. Jika ada, kita tambahkan 1 ke true_pos. Akhirnya, kita mengembalikan hasil pembagian float dari true_pos dengan panjang model_output untuk mendapatkan nilai precision.

Fungsi calculate_found digunakan untuk menentukan apakah ada item yang ditemukan dalam model_output yang juga ada dalam gold_standard. Fungsi ini mengembalikan nilai 1 jika ada item yang cocok dan 0 jika tidak ada.

Kedua fungsi ini membandingkan model_output dengan gold_standard dan memberikan informasi tentang kinerja model dalam mengidentifikasi item yang benar.

```python
def calculate_precision(model_output, gold_standard):
    true_pos = 0
    for item in model_output:
        if item in gold_standard:
            true_pos += 1
    return float(true_pos)/float(len(model_output))

def calculate_found(model_output, gold_standard):
    found = 0
    for item in model_output:
        if item in gold_standard:
            found = 1
    return float(found)
```

```python
import torch

# Convert query vectors and document vectors to PyTorch tensors
qry_tensors = {}
for query_id, query_vector in qry_vectors.items():
    qry_tensors[query_id] = torch.Tensor(query_vector)

doc_tensors = {}
for doc_id, doc_vector in doc_vectors.items():
    doc_tensors[doc_id] = torch.Tensor(doc_vector)

# Create a CUDA device for running computations on the GPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Move query and document tensors to the device
for query_id, qry_tensor in qry_tensors.items():
    qry_tensors[query_id] = qry_tensor.to(device)

for doc_id, doc_tensor in doc_tensors.items():
    doc_tensors[doc_id] = doc_tensor.to(device)

# Calculate cosine similarities between query vectors and document vectors
precision_all = 0.0
found_all = 0.0
for query_id in qrel_list.keys():
    gold_standard = qrel_list[query_id]
    query = qry_tensors[query_id].unsqueeze(0)  # Add batch dimension
    results = {}
    model_output = []
    for doc_id in doc_tensors.keys():
        document = doc_tensors[doc_id].unsqueeze(0)  # Add batch dimension
        cosine = torch.nn.functional.cosine_similarity(query, document).item()
        results[doc_id] = cosine
    for items in sorted(results.items(), key=lambda x: x[1], reverse=True)[:30]: #JUMLAH RANKING RESULT, MENGGUNAKAN 30 UNTUK MENYAMAI CODE NILEYGF
        model_output.append(items[0])
    precision = calculate_precision(model_output, gold_standard)
    found = calculate_found(model_output, gold_standard)

    #only uncomment which one you wanted to use!!

    print(str(query_id) + ": " + str(precision))    #print precision score with query id (main line)
    #print(str(precision) + ", ")                    #print precision score with comma seperation (optional)
    #print(str(found) + ", ")                         #print found score with comma seperation (optional)

    precision_all += precision
    found_all += found
```
